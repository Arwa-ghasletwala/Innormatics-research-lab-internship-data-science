*TASK 1*   
DEFANGING ON IP ADDRESS

```python
def defangIP(address):
    output=""
    for i in address:
        if i=='.':
            output+='[.]'
        else:
            output+=i
    return output
```

*TASK 2*   
FIND NUMBERS WITH AN EVEN NUMBER OF DIGITS

```python
def even_number_of_digits(nums):
    count=0
    for i in nums:
        temp=0
        while i>0:
            i=i//10
            temp+=1
        if temp%2==0:
            count+=1
    return count 
```

*TASK 3*   
NUMBER OF GOOD PAIRS

```python
def identical_pairs(nums):
    count=0
    for i in range(len(nums)):
        for j in range(i,len(nums)):
            if nums[i]==nums[j] and i<j:
                count+=1
    return count
```

*TASK 4*   
HOW MANY NUMBERS ARE SMALLER THAN CURRENT NUMBER

```python
def samller_than_current(nums):
    count=[]
    for i in nums:
        temp=0
        for j in range(len(nums)):
            if i>nums[j]:
                temp+=1
        count.append(temp)
    return count
```

*TASK 5*
SUBTRACT THE PRODUCT AND SUM OF DIGITS OF AN INTEGER

```python
def sub_product_and_sum(num):
    product=1
    sum=0
    while num>0:
        temp=num%10
        product*=temp
        sum+=temp
        num=num//10
    return product-sum
```

*TASK 6*
XOR OPERATION IN AN ARRAY

```python
def XOR(n,start):
    nums=[]
    for i in range(n):
        nums.append(start+2*i)
    xor_operation=nums[0]
    for i in range(1,n):
        xor_operation=xor_operation^nums[i]
    return xor_operation
```
