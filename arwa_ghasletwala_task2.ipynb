__#TASK 1__  
Given an array nums. We define a running sum of an array as runningSum[i] = sum(nums[0]â€¦nums[i]).  
Return the running sum of nums.

```python
def running_sum(nums):
    output=[]
    for i in range(len(nums)):
        sum=0
        for j in range(i+1):
            sum+=nums[j]
        output.append(sum)
    return output
nums=[]
n=int(input("array length :"))
print()
for i in range(n):
    a=int(input("enter number :"))
    nums.append(a)
print("running sums array :",running_sum(nums))
```

__#TASK 2__  
Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].  
Return the array in the form [x1,y1,x2,y2,...,xn,yn].

```python
def shift_array(nums,n):
    output=[]
    for i in range(n):
        output.append(nums[i])
        output.append(nums[i+n])
    return output
nums=[]
size=int(input("array length (even numebr) :"))
print()
for i in range(size):
    a=int(input("enter number :"))
    nums.append(a)
print("original array :",nums)
n=int(size/2)
print("shifted array :",shift_array(nums,n))
```

__#TASK 3__  
There are n kids with candies. You are given an integer array candies, where each candies[i] represents the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.  
Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, they will have the greatest number of candies among all the kids, or false otherwise.

```python
def maximum_candies(candies,extra_candy):
    output=[]
    max_candies=max(candies)
    new_candies=[x + extra_candy for x in candies]
    for i in new_candies:
        if i>=max_candies:
            flag=True
        else:
            flag=False
        output.append(flag)
    return output
candies=[]
size=int(input("enter number of kids :"))
print()
for i in range(size):
    a=int(input("enter number candies :"))
    candies.append(a)
extra_candy=int(input("enter number of extra candy :"))
print("greatest number of candies or not :",maximum_candies(candies,extra_candy))
```
